```python
import json
conteo = {}
with open('SABANAS_INE_ELECCION_2012/file_id-0.json') as file:
    data = json.load(file)
    for elemento in data['sabanas']:
        for partido in elemento['resultados']:
            print(partido, elemento['resultados'].get(partido))
            conteo[partido] =  conteo.get(partido,0) +  int(elemento['resultados'].get(partido))
print (conteo)
```

```python
conteo = {}
with open('SABANAS_INE_ELECCION_2012/nombre_archivos.txt','r') as archivo:
    for nombre_archivo in archivo:
        nombre_archivo = nombre_archivo.strip()
        with open('SABANAS_INE_ELECCION_2012/'+nombre_archivo) as file:
            data = json.load(file)
            for elemento in data['sabanas']:
                for partido in elemento['resultados']:
                    
                    conteo[partido] =  conteo.get(partido,0) +  int(elemento['resultados'].get(partido))
print (conteo)
```
