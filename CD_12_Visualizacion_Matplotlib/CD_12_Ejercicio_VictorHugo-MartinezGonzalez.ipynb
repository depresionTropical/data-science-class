# Ejercicio 2: Incendios en México

```python
df= pd.read_csv("datos/Serie_historica_anual_incendios_2017.csv")
```

```python
df.set_index('Fecha Inicio', inplace=True)
df.index = pd.to_datetime(df.index , format='%d/%m/%Y')
df
```

<div class="alert alert-success">
    
1. Utilizar una grafica de barras para comparar el total de hectáreas afectadas para los 5 estados más afectados 
    
</div> 

```python
# Agrupar por estado y sumar las hectáreas afectadas
hectareas_por_estado = df.groupby('Estado')['Total Hectareas'].sum()

# Seleccionar los 5 estados más afectados
top_5_estados = hectareas_por_estado.nlargest(5)

# Crear la gráfica de barras
plt.bar(top_5_estados.index, top_5_estados.values,color='skyblue')
plt.xlabel('Estado')
plt.ylabel('Total de Hectáreas Afectadas')
plt.title('Total de Hectáreas Afectadas para los 5 Estados Más Afectados')
plt.show()
```

<div class="alert alert-success">
    
2. Utilizar una grafica de barras para comparar el total de hectáreas afectadas para Yucatán para los meses: abril, mayo, junio, julio y agosto 
    
</div> 

```python
df_yucatan = df[(df.index.month.isin([4,5,6,7,8]))&(df.Estado.isin(['Yucatán']))]['Total Hectareas'].resample('M').sum()
df_yucatan
meses = df_yucatan.index.strftime('%B')
plt.bar(meses, df_yucatan.values, color='red')
plt.xlabel('Mes')
plt.ylabel('Total de Hectáreas Afectadas')
plt.title('Total de Hectáreas Afectadas por Mes en Yucatán')
plt.xticks(rotation=45)
plt.show()
```

<div class="alert alert-success">
    
3. Considerando todos los datos, hacer un histograma de la duración de los incendios 
    
</div> 

```python
df['Duracion H'].dropna().max()
```

```python
plt.hist(df['Duracion H'].dropna(), bins=5, edgecolor='black', color='skyblue', alpha=0.7)
plt.xlabel('Duración de los Incendios (Horas)')
plt.ylabel('Frecuencia')
plt.title('Histograma de la Duración de los Incendios')
plt.show()
```

```python
import numpy as np
plt.hist(np.log1p(df['Duracion H'].dropna()), bins=20, edgecolor='black', color='skyblue', alpha=0.7)
plt.xlabel('Log de la Duración de los Incendios (log(Horas + 1))')
plt.ylabel('Frecuencia')
plt.title('Histograma de la Duración de los Incendios (Transformación Logarítmica)')
plt.show()
```

<div class="alert alert-success">
    
4. Considerando todos los datos, hacer una grafica de dispersión entre la duración (eje X) y el total de hectáreas (eje Y)
    
</div> 

```python
plt.scatter(df['Duracion H'], df['Total Hectareas'], alpha=0.5)  # Ajusta el valor de 's' según sea necesario
plt.xlabel('Duración de los Incendios (Horas)')
plt.ylabel('Total de Hectáreas Afectadas')
plt.title('Dispersión entre la Duración de los Incendios y el Total de Hectáreas Afectadas')
plt.show()
```
