<div class="alert alert-success">

<b>EJERCICIO 1</b>:

Explicar las opciones Inner,Left, Right y Outer para la función _merge_, y muestre 4 ejemplos de su uso

</div>


```python
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],
                    'value1': [1, 2, 3, 4]})

df2 = pd.DataFrame({'key': ['B', 'D', 'E', 'F'],
                    'value2': [5, 6, 7, 8]})
df1
```

```python
df2
```

```python
# ejemplo 1
# Solo incluye los datos que están presentes en ambas tablas en las columnas clave.
pd.merge(df1, df2, on='key', how='inner')
```

```python
# ejemplo 2
# Mantiene todos los datos del DataFrame de la izquierda.
pd.merge(df1, df2, on='key', how='left')
```

```python
# ejemplo 3
# Mantiene todos los datos del DataFrame de la derecha.
pd.merge(df1, df2, on='key', how='right')
```

```python
# ejemplo 4
# Combina todos los datos, incluso aquellos que no tienen coincidencias.
pd.merge(df1, df2, on='key', how='outer')
```
